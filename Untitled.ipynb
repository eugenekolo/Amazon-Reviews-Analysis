```python
#!/usr/bin/env python

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import seaborn as sns
import os
from collections import defaultdict

from textblob import TextBlob

import gzip
import json
import time
import datetime
import dateutil
%matplotlib inline

from nltk.tokenize import RegexpTokenizer
from nltk import word_tokenize

# Configure the analysis
BEAUTY_REVIEWS_FILE = "res/reviews_Beauty.json.gz" # Beauty (2,023,070 reviews)
BABY_REVIEWS_FILE = "res/reviews_Baby.json.gz" # Baby (915,446 reviews)
ELEC_REVIEWS_FILE = "res/reviews_Electronics-small.json.gz" # Electronics Subset (781,889 reviews)
ELEC_REVIEWS_FILE_FULL = "res/reviews_Electronics.json.gz"
REVIEWS_FILE = ELEC_REVIEWS_FILE_FULL
```

```python
# Load the review data
def json_from_gzip_file(file_name):
    reviews = []
    with gzip.open(file_name) as gz_file:
        for jsonstr in gz_file:
            review = json.loads(jsonstr.decode('utf-8'))
            reviews.append(review)

    return reviews

def list_to_pd(mylist):
    return pd.DataFrame(mylist)

def gzip_json_to_pd(file_name):
    list_to_pd(json_from_gzip_file(file_name))
    

# Parts of a review:
#     reviewerID     - ID of the reviewer, e.g. A2SUAM1J3GNN3B
#     asin           - ID of the product, e.g. 0000013714
#     reviewerName   - name of the reviewer
#     helpful        - helpfulness rating of the review, e.g. 2/3
#     reviewText     - text of the review
#     overall        - rating of the product
#     summary        - summary of the review
#     unixReviewTime - time of the review (unix time)
#     reviewTime     - time of the review (raw)

reviews_jsons = []
print("[*] Loading %s..." % REVIEWS_FILE)

reviews = json_from_gzip_file(REVIEWS_FILE)

print("[*] Loaded %s" % REVIEWS_FILE)
reviews = list_to_pd(reviews)

print("[*] Created DataFrame from JSON objects")  
```

```python
# Shape the data for what we're interested in

# Find reviews that are voted helpful, and their percent_helpful
MIN_HELPFUL_VOTES = 3
PERCENT_TO_BE_HELPFUL = .7

print("[*] Cleaning up the data...")
reviews['helpful_votes'] = reviews.helpful.apply(lambda x: x[0])
reviews['total_votes'] = reviews.helpful.apply(lambda x: x[1])

reviews['percent_helpful'] = reviews['helpful_votes'] / reviews['total_votes']
reviews['percent_helpful'] = reviews['percent_helpful'].fillna(value=.5)
reviews["percent_helpful_scaled"] = 2*((reviews["percent_helpful"] - 0)/(1 - 0)) - 1 # Normalize to ()

reviews['is_helpful'] = np.where((reviews.percent_helpful > PERCENT_TO_BE_HELPFUL) &
                                 (reviews.helpful_votes > MIN_HELPFUL_VOTES), "Yes", "No")
reviews['year_written'] = reviews.unixReviewTime.apply(lambda x: time.gmtime(x)[0])
# Throw out incomplete years for better analysis
reviews = reviews[(reviews.year_written > 1996) & (reviews.year_written < 2014)]
#reviews.head()
#reviews_sample = reviews.sample(frac=.1)
```

```python
#getting more features

###feature : reviewers average rating
#getting reviewers average rating and appending it to the df
reviewers_avg_rating = reviews.groupby("reviewerID", as_index=False)['overall'].mean()
reviews = pd.merge(reviews, reviewers_avg_rating, on="reviewerID")
reviews.rename(columns = {'overall_y':'average_rating'}, inplace=True)
print("[*] Done adding reviewers average rating feature...")



### feature: max_reviews_on_any_day  <--- takes a long time
group_by_id_time = reviews.groupby(["reviewerID", "reviewTime"])["reviewerID"].agg({'overall_x': 'count'})
mask = group_by_id_time.groupby(level=0).agg('idxmax')
df_count = group_by_id_time.loc[mask['overall_x'].tolist()]
df_count = df_count.reset_index()
df_count.rename(columns={"overall_x" : "max_reviews_any_day"}, inplace=True)
df_count.drop("reviewTime", axis=1, inplace=True)
reviews = pd.merge(reviews, df_count, on="reviewerID")
reviews.rename(columns={"overall_x": "overall"}, inplace=True)
print("[*] Done adding max reviews on any day for a reviewer...")

#### feature: total number of reviews given by user 
group_by_id = reviews.groupby("reviewerID").count()
group_by_id["reviewerID"] = group_by_id.index
group_by_id = group_by_id[["overall", "reviewerID"]]
group_by_id.rename(columns={"overall_y" : "total_num_reviews"}, inplace=True)
reviews = pd.merge(reviews, group_by_id, on="reviewerID")
reviews.rename(columns={"overall_y" : "total_num_reviews"}, inplace=True)
print("[*] Done adding total number of reviews for a reviewer...")

# #### feature: reviews per day 
reviews["reviews_per_day"] = reviews["total_num_reviews"] / 365.0
print("[*] Done adding reviews per day...")

### feature: number of words in review text (some text is off by 1 or 2 words, not sure whyw)
reviews['review_length'] = reviews['reviewText'].str.split().apply(len)
print("[*] Done adding number of words in review text...")

#fixing column names:
#reviews.rename(columns={"overall_x":"overall"})
```

```python
reviews['polarity'] = reviews['reviewText'].apply(lambda x: TextBlob(x).sentiment.polarity)
print("[*] Done adding polarity...")
reviews['subjectivity'] = reviews['reviewText'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
def jaccard_similarity(a, b):
    """Calculate the Jaccard Similarity of two sets.
    https://en.wikipedia.org/wiki/Jaccard_index
    The Jaccard index, also known as the Jaccard similarity coefficient (originally coined
    coefficient de communauté by Paul Jaccard), is a statistic used for comparing the similarity
    and diversity of sample sets. The Jaccard coefficient measures similarity between finite sample
    sets, and is defined as the size of the intersection divided by the size of the union of the
    sample sets:

    Args:
        a:  First set to compare
        b:  Second set to compare to

    Returns:
        A similarity score between 0 and 1.
        0 means no similarity, and 1 means perfect similarity.
    """

    size_of_intersection = len(set.intersection(a, b))
    size_of_union = len(set.union(a, b))

    similarity =  size_of_intersection / float(size_of_union)

    return similarity

def cosine_similarity(a, b):
    pass

def calculate_similarity(a, b, func):
    return func(a,b)


review_text_by_reviewer = reviews.groupby("reviewerID")

review_text_by_reviewer['reviewText'].apply()

#for name, group in review_text_by_reviewer.head(100):
#    print(name, group)
    
```

```python
# Examine ratio of helpful reviews to total reviews year to year
net_reviews_good = reviews[reviews['is_helpful'].isin(['Yes'])]['year_written'].value_counts()
net_reviews_bad = reviews[reviews['is_helpful'].isin(['No'])]['year_written'].value_counts()
net_reviews_ratio = net_reviews_good / (net_reviews_bad + net_reviews_good)
ax = sns.barplot(net_reviews_ratio.index, net_reviews_ratio)
ax.set_title('Ratio of Helpful Reviews To Total Reviews Year to Year')
ax.set_xlabel('Year')
ax.set_ylabel('Ratio')
plt.show()
```
